```python
# RMSE Function
# RMSE = sqrt(sum((predicted_i - actual_i)^2)/total number of predictions)

import math

def RMSE(predicted, actual):
    sums = 0.0
    for i in range(len(predicted)):
        difference = predicted[i]- actual[i]
        squared_difference = difference**2
        sums += squared_difference
        return sums
    MSE = sums/len(predicted)
    return math.sqrt(MSE)
```

```python
# The data
x = [0, 1, 2, 3, 4]
y = [0.1, 0.9, 2.1, 2.9, 4.1]

y_1 = [0, 1, 2, 3, 4]
y_2 = [0.1, 0.9, 2.1, 2.9, 4.1]

#(x,y) = (5, 5.1)
#y_1 = 5
#y_2 = 5.2
```

```python
print('Root Mean Squared Error of Model_1:')
RMSE(y_1,y)
```

```python
print('Root Mean Squared Error of Model_2:')
RMSE(y_2, y)
```

_Comments:_  When we look at the RMSE values for the two given models, it seems like the model_2 is the perfect model because it returns 0.0 RMSE value. Even though the lower RMSE the better, it looks suspicious. So, let's take a look at the graphs below to compare two models.

```python
# Libraries and modules
import matplotlib.pyplot as plt

# Figures
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x, y, color = 'red')
plt.plot(x,y_1, color = 'blue', linewidth = 2)
plt.ylabel('y')
plt.xlabel('x')
plt.title('model_1')

plt.subplot(1,2,2)
plt.scatter(x, y, color = 'red')
plt.plot(x,y_2, color = 'blue', linewidth = 2)
plt.ylabel('y')
plt.xlabel('x')
plt.title('model_2')
```

_Comments:_ When we compare the two models the first one looks like an almost perfect linear regression model. And the model_2 looks like it has an __overfitting__ problem where it possibly will not give good results on an unseen test data. 

If we consider the outcomes of the test point (x,y) = (5, 5.1);
- The first model returns exactly the expected value, which is 5
- The second model should return 5.1 but instead it returns 5.2 which proves the overfitting problem 

I would say the __model_1__ is a better model because of the explanation I gave above.
